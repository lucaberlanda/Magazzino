# <span style='font-family:Georgia'>  Log-log vs semi-log plot

- **All equations of the form: $y = a^x$ form a straight lines when plotted semi-logarithmical graph**. Taking the log on both sides,   
    
    $$
    \log_{a}(y) = \log_{a}(a^x)
    $$

    hence, $\log_{a}(y) = x + \log(c)$; setting $\log(c) = k$, we obtain $\log_{a}(y) = x + k$; then is a linear function of $x$


- **Relations of the form $y= x^k$ appear as straight lines in a logâ€“log graph**. Taking the log on both sides, 
  
  $$
  \log(y) = \log(x^\alpha)
  $$
  
  hence, $\log(y) = \alpha \log(x)$; setting $\log(y) = Y$ and $\log(x) = X$, we obtain $Y = \alpha X$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## <span style='font-family:Georgia'> Power Law

```python
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

exponent = 2
xs = pd.DataFrame(np.arange(1, 100))
ys = xs**exponent

ax.scatter(xs, ys, alpha=0.8, c='#5514ED')
ax.set_yscale('log')
ax.set_xscale("log")
ax2.scatter(xs, ys, alpha=0.8, c='#5514ED')

ax.set_title('Log Log Plot', fontsize=18)
ax2.set_title('Raw Data', fontsize=18)

plt.show()
```

## <span style='font-family:Georgia'>  Exponential Distribution

```python
fig2 = plt.figure(figsize=(14, 6))
ax3 = fig2.add_subplot(121)
ax4 = fig2.add_subplot(122)

base = 2
xs2 = pd.DataFrame(np.arange(1, 30))
ys2 = base ** xs2

ax3.scatter(xs2, ys2, c='#F72585')
ax3.set_yscale('log')
ax4.scatter(xs2, ys2, c='#F72585')

ax3.set_title('Semi Log Plot', fontsize=18)
ax4.set_title('Raw Data', fontsize=18)

plt.show()
```
