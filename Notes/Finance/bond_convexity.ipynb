# <span style='font-family:Inter'> Bond Convexity

```python
import numpy as np
from Finance.Bond.bond_calculator import Bond

yield_list = list(np.arange(-0.01, 0.4, 0.01))

price_yield = {}
for ytm in yield_list:
    bond = Bond(face_value=100, 
                coupon=0.03,
                years=10,
                price=None,
                ytm=ytm)
    
    price_yield[ytm] = bond.get_metrics()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
p_y = pd.Series(price_yield)
p_y.plot(ax=ax)
ax.set_xlabel("yield")
ax.set_ylabel("price")
ax.set_title('Bond Convexity', fontsize=15)
plt.show()
```
