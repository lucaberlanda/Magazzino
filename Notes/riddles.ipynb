## <span style='font-family:Inter'> Absorbing Barrier

[Tweet](https://twitter.com/mkeharrisNY/status/1315426197413621760)

> Sunday probability puzzle: Three fair coins tossed. Heads gets +1, tails -1, pay-offs are added and net pay-off added to equity. The 3 tosses are repeated 1000 times. Initial equity is 10 dollars. What is the probability of total ruin (within +/- 0.05 error)? Please do not post answers.

```python
np.random.seed(100)
t = 1000
samples = 1000

df_dict = {}
for cnt2 in range(1, samples):
    money = 10
    money_dict = {}
    for i in range(1, t):
        if money == 0:
            pass
        else:
            money += sum(np.random.choice([-1, 1], 3))
            money_dict[i] = max(money, 0)

    df_dict[cnt2] = pd.Series(money_dict)
```

```python
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(1, 1, 1)

absorbed_proportion = (pd.DataFrame(df_dict).fillna(0) == 0).mean(axis=1)
absorbed_proportion.plot(ax=ax, color='#3A0CA3')
ax.set_title('Proportion of Absorbed Paths', fontsize=15)

plt.show()
HTML(center_string)
```

```python
sorted_df = pd.DataFrame(df_dict).loc[:, pd.DataFrame(df_dict).iloc[-1, :].fillna(0).sort_values().index]
sorted_df.plot(legend=False, figsize=(10,7), cmap='inferno', alpha=0.5)
plt.show()
```

## <span style='font-family:Inter'> Consecutive Heads

number of consecutive heads or tail in $n$ draws


```python
import numpy as np
n = 100

summation = 0
for k in np.arange(1, n/2 + 1):
    print(k, 2 ** (-(n-k)))
    summation += 2 ** (-(n-k))
    
print(summation)
exp_cons_h_t = (n + 1) * summation
```

## <span style='font-family:Inter'> 20 Balls More

[Link](http://www.qbyte.org/puzzles/p064s.html)

```python
for k in range(2, 20):  # colors
    for n in range(2, 20):  # n balls of same colors
        case1 = (n - 1)/(n*k -1)
        case2 = ((n-1)/(n*k  + 19))*((n*k)/(n*k  + 20)) + ((19)/(n*k  + 19))*(20/(n*k  + 20)) # old balls + new balls
        if round(case1 * 100, 4) == round(case2 * 100, 4):
            print('n =', n,', k =', k)
```
